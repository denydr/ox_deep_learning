# Decision Trees and Random Forests

Decision trees are hailed as one of the easiest-to-understand concepts in machine learning. They involve starting at a root and making a series of decisions until an answer is reached. Unfortunately, their simplicity also makes them inaccurate: they tend to overfit the data, and thus, they are not wildly used in machine learning.

Random forests improve decision trees dramatically. They involve using multiple decision trees in a decision, as well as several randomization procedures to combat overfitting. In the process, they combine the simplicity of the decision tree with the flexibility of more complex machine learning models.

In this chapter, we will start off with a decision tree, and then we will advance to a random forest model

## 3 | Fit a Decision Tree via Sklearn

### 3.1 | Fit Simple Model (i.e., one with no hyperparameters)

Fitting a decision tree can become quite complicated. To avoid this complication, we will initially fit a decision tree without most parameters. The only one we will specify is random_state, which seeds Sklearn's random number generator. In other words, it ensures that everyone's random number generator functions in the same way, thus leading to everyone getting the same result.

```python
#Fits a decision tree via three lines of code
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
```

We fitted our first decision tree with the training data! However, we need some way of evaluating it. Let's do that below.

```python
#Function to facilitate evaluating our models
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```

```python
# Print results of our evaluation
print_score(tree_clf, X_train, y_train, X_test, y_test, train=True)
print_score(tree_clf, X_train, y_train, X_test, y_test, train=False)
```

```python
#Let's visualize the confusion matrix via seaborn
import seaborn as sns
from sklearn.metrics import confusion_matrix
 
y_hat = tree_clf.predict(X_test) > 0.5 #Note that, in machine learning lingo, we call predicted y values 'y hat' which is written as ŷ
mat = confusion_matrix(y_test, y_hat)
labels = ['No Homicide', 'Homicide']
 
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=labels, yticklabels=labels)
 
plt.xlabel('Predicted label')
plt.ylabel('True label')
```

Let's briefly analyze the confusion matrix.

Our *accuracy score* for the testing set is .8580! In other words, it correctly labelled 85.80% of our datapoints! Our model must be amazing!

Not quite. There are 51 homicides in our dataset, yet our decision tree could only predict just 8 of them; it incorrectly labelled the rest (n=43) as a false negative. Thus, our *sensitivity score (recall score)* is 8/(8+43)=.157. This isn't the same, ultra-high accuracy score of 0.8580. Thus, our model isn't great at predicting homicide.

There's also a ton of false positives. For every 36 people flagged by our model as being a homicide victim, only 8 of them will actually become victims, 28 will not. In other words, let's imagine that we're using this algorythm to predict who to arrest: we'll arrest the partner of any individual who our model thinks will become a victim of domestic homicide. If that's the case, then for every 2 people whose lives we'll save, we'll be possibly ruining the lives of 7 other people, locking up their partner for domestic homicide when they would have committed the crime! This is not an acceptable ratio. (In machine learning terms, our *percision* score would be 8/(8+28)= 0.22. This is not the same high .8580 score as our accuracy.)

Wait! We can fine-tune a machine learning model - make it even better - by doing something called hyperparameter tuning. Let's try that now and make a better model.

### 3.2 | Hyperparameter Tuning

We'll discuss hyperparameter tuning later. For now, just know that it involves running hundreds of emperical tests to pick out the absolute best values for our model. It should lead to a significant performance increase.

```python
#Don't worry about the code. Run it, and we will discuss it later.
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

params = {
    "criterion":("gini", "entropy"), 
    "splitter":("best", "random"), 
    "max_depth":(list(range(1, 20))), 
    "min_samples_split":[2, 3, 4], 
    "min_samples_leaf":list(range(1, 20)), 
}

tree_clf = DecisionTreeClassifier(random_state=42)
tree_cv = GridSearchCV(tree_clf, params, scoring="accuracy", n_jobs=-1, verbose=1, cv=3)
tree_cv.fit(X_train, y_train)
best_params = tree_cv.best_params_
print(f"Best paramters: {best_params})")
```

```python
#Refits model
tree_clf = DecisionTreeClassifier(**best_params)
tree_clf.fit(X_train, y_train)

#Prints scores 
print_score(tree_clf, X_train, y_train, X_test, y_test, train=False)

#Fits confusion matrix
y_hat = tree_clf.predict(X_test) > 0.5 
mat = confusion_matrix(y_test, y_hat)
labels = ['No Homicide', 'Homicide']
 
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=labels, yticklabels=labels)
 
plt.xlabel('Predicted label')
plt.ylabel('True label')
```

After tuning the hyperparameters, our decision tree did *even worse* than before! Our accuracy score shot up, from .8580 to 0.92.

However, our model has only successfully predicted *a single homicide*, and the rest were all false positives! In other words, if one were to take the test dataset and predict that absolutely no-one in it will commit a domestic homicide, then that guess would be better than the performance of the model.

Why is our optimized model so terrible at predicting homicide? Let's return to our data.

## 4 | Dataset Revisited

```python
#First, Let's have a look at the labels again. 
# We'll only look at the training set because that's the dataset that the model is training on --- the same dataset that is causing the model to behave poorly.

y_train.value_counts()
```

Our training dataset is wildly unbalanced. This lack of balance is very dangerous for machine learning because models learn that, if they ignore the minority case, they can achieve a very high accuracy score. 

For example, imagine a dataset where there's 9,900 individuals who will never commit a homicide, and 100 individuals who will. Machine learning models will learn that they can achieve a superb .99 accuracy score by *ignoring* those who will commit a homicide. This is extremely problematic because the machine is ignoring the same case that we're most interested in predicting.

To fix this lack of balance, the best case scenario is to get more data. However, if that is not possible, then one can "trick" the machine into caring about the minority case via a variety of techniques. We can undersample the majority case (i.e. individuals who will never commit a homicide), but this isn't ideal because we'd be throwing away data. We can oversample homicide data, but this isn't ideal either because we'd be feeding our model redundant data. We can also change the hyperparameter tuning procedure to focus on percision, a metric designed for unbalanced data, but this typically does not improve results by too much.

Instead of the above three procedures, the pocedure we'll instead employ is **artificial data generation via SMOTE**, an algorithm that generates data via interpolation. We can use this to artificially generate homicide data, thus making the dataset more balanced. 

Using SMOTE by itself is not perfect for reasons discussed in this lecture. Ideally, a combination of the procedures will be used. However, this is a quick fix and will be sufficient for the purpose of this lesson.

## 5 | Artificial Data Generation via SMOTE

As discussed in the above section, using SMOTE to fix an unbalanced dataset is imperfect. Ideally, one would gain more data on the minority case, and if that is not possible, then a variety of other techniques will be used in addition to SMOTE. However, SMOTE will be sufficient for this lesson.

```python
#Homicide data before SMOTE
y_train.value_counts()
```

```python
#Apply SMOTE to the training set.
from imblearn.over_sampling import SMOTE
smote_resampler = SMOTE()
X_train, y_train = smote_resampler.fit_resample(X_train, y_train)

#Prints new dataset
y_train.value_counts()
```

Much better. Now that the dataset is balanced, let's return to the decision tree.

## 6 | Fit a Decision Tree, Revisited

### 6.1 | Fit Simple Model (i.e., one with no hyperparameters)

```python
#Fits a decision tree 
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)

#Prints scores 
print_score(tree_clf, X_train, y_train, X_test, y_test, train=False)

#Plots confusion matrix
y_hat = tree_clf.predict(X_test) > 0.5 
mat = confusion_matrix(y_test, y_hat)
labels = ['No Homicide', 'Homicide']
 
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=labels, yticklabels=labels)
 
plt.xlabel('Predicted label')
plt.ylabel('True label')
```

Much better! Our decision tree is still far from ideal, but it is predicting more homicides than the previous two attempts. 

Let's return to hyperparameter tuning to really improve our decision tree.

### 6.2 | Hyperparameter Tuning

We can achieve a better performing decision tree by specifying the parameters. These are once again called hyerparameters in machine learning.

However, there are many parameters, and we do not know which values will lead to the best performance! Indeed, each time the dataset is changed, the best performing values may also change.

To solve this problem, we engage in a process known as hyperparameter tuning (sometimes called "hyperparameter optimization"). This approach plugs in a bunch of values for these parameters, and it finds which combination of parameters leads to the best output.

```python
#These are the parameters we're specifying, as well as the values we're testing
params = {
    "criterion":("gini", "entropy"), 
    "splitter":("best", "random"), 
    "max_depth":(list(range(1, 20))), 
    "min_samples_split":[2, 3, 4], 
    "min_samples_leaf":list(range(1, 20)), 
}
```

However, know that this is a subset of all the possible hyperparameters that we could be specifying. Each hyperparameter and its description appears below, adapted from the sklearn documentation.

* criterion: The function to measure the quality of a split. Supported criteria are "gini" for the Gini impurity and "entropy" for the information gain.

* splitter: The strategy used to choose the split at each node. Supported strategies are "best" to choose the best split and "random" to choose the best random split.

* max_depth: The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

* min_samples_split: The minimum number of samples required to split an internal node.

* min_samples_leaf: The minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. This may have the effect of smoothing the model, especially in regression.

* min_weight_fraction_leaf: The minimum weighted fraction of the sum total of weights (of all the input samples) required to be at a leaf node. Samples have equal weight when sample_weight is not provided.

* max_features: The number of features to consider when looking for the best split.

* max_leaf_nodes: Grow a tree with max_leaf_nodes in best-first fashion. Best nodes are defined as relative reduction in impurity. If None then unlimited number of leaf nodes.

* min_impurity_decrease: A node will be split if this split induces a decrease of the impurity greater than or equal to this value.

* min_impurity_split: Threshold for early stopping in tree growth. A node will split if its impurity is above the threshold, otherwise it is a leaf.

Let's test these parameters to see which values work.

```python
#Hyperparameter tuning. It's recommended to open Anoconda so that you can see the model testing these hyperparameters in real time.
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV


tree_clf = DecisionTreeClassifier(random_state=42)
tree_cv = GridSearchCV(tree_clf, params, scoring="accuracy", n_jobs=-1, verbose=3, cv=3)
tree_cv.fit(X_train, y_train)
best_params = tree_cv.best_params_
print(f"Best paramters: {best_params})")
```

```python
#Re-fit the decision tree using the best hyperparameters
tree_clf = DecisionTreeClassifier(**best_params)
tree_clf.fit(X_train, y_train)
```

```python
#Let's see how the new, tuned model performed.
print_score(tree_clf, X_train, y_train, X_test, y_test, train=False)

#Plots confusion matrix
y_hat = tree_clf.predict(X_test) > 0.5 
mat = confusion_matrix(y_test, y_hat)
labels = ['No Homicide', 'Homicide']
 
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=labels, yticklabels=labels)
 
plt.xlabel('Predicted label')
plt.ylabel('True label')
```

The hyperparameter tuned decision tree performs noticeably better than its original counterpart: it has less false postives, less false negatives, and more true positives than its counterpart.

Recall that, before the SMOTE algorithm, hyperparameter tuning resulted in a *worse* dataset. This is no longer the case.

However, we can do substantially better via a random forest model, to which we now turn our attention to.

## 7 | Random Forest

### 7.1 | Simple Fitting

We'll now fit a random forest model to our dataset. We'll first fit the model with minimal hyperparameters, much like we did for decision trees.

```python
from sklearn.ensemble import RandomForestClassifier

#Fits random forest model
rf_clf = RandomForestClassifier(n_estimators=100)
rf_clf.fit(X_train, y_train)

#Evaluates model
print_score(rf_clf, X_train, y_train, X_test, y_test, train=False)

#Plots confusion matrix
y_hat = rf_clf.predict(X_test) > 0.5 
mat = confusion_matrix(y_test, y_hat)
labels = ['No Homicide', 'Homicide']
 
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=labels, yticklabels=labels)
 
plt.xlabel('Predicted label')
plt.ylabel('True label')
```

We did not even optimize the random forest model, yet it is already performing better than the optimized decision tree from §6.2 in every metric (i.e., specificity, sensitivity, and precision). However, we can do better after hyper-tuning it.

### 7.2 | Hyperparameter Tuning

```python
#These are the hyperparameters that we'll be tunning
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=10, stop=1000, num=10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num=11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

grid_search = {'n_estimators': n_estimators, 'max_features': max_features,
               'max_depth': max_depth, 'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf, 'bootstrap': bootstrap}
```

Just like decision trees, the hyperparameters used above are a subset of all hyperparameters. All the hyperparameters, as well as a description of each, appears below, adapted from the Sklearn documentation. 

* n_estimators: The number of trees in the forest.
* criterion: The function to measure the quality of a split. Supported criteria are "gini" for the Gini impurity and "entropy" for the information gain.
* max_depth: The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
* min_samples_split: The minimum number of samples required to split an internal node.
* min_samples_leaf: The minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. This may have the effect of smoothing the model, especially in regression.
* min_weight_fraction_leaf: The minimum weighted fraction of the sum total of weights (of all the input samples) required to be at a leaf node. Samples have equal weight when sample_weight is not provided.
* max_features: The number of features to consider when looking for the best split.
* max_leaf_nodes: Grow a tree with max_leaf_nodes in best-first fashion. Best nodes are defined as relative reduction in impurity. If None then unlimited number of leaf nodes.
* min_impurity_decrease: A node will be split if this split induces a decrease of the impurity greater than or equal to this value.
* min_impurity_split: Threshold for early stopping in tree growth. A node will split if its impurity is above the threshold, otherwise it is a leaf.
* bootstrap: Whether bootstrap samples are used when building trees. If False, the whole datset is used to build each tree.
* oob_score: Whether to use out-of-bag samples to estimate the generalization accuracy.

```python
#Tests the hyperparameters. You can view the testing in real time by looking at your anoconda window.
#Beecause a random forest has a lot more hyerparameters to optimize than a decision tree, a random grid search (vs. grid search) is used.
#The random grid search doesn't test all hyperparameter possibilities: it only tests a subset of them. However, that subset should be sufficient to produce great results.
rf_clf = RandomForestClassifier(random_state=42)
rf_cv = RandomizedSearchCV(estimator=rf_clf, scoring='accuracy',param_distributions=grid_search, n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1)


#Fits data 
rf_cv.fit(X_train, y_train)

#Finallly, print out the best results 
rf_best_params = rf_cv.best_params_
print(f"Best paramters: {rf_best_params})")
```

```python
#Fits the model and prints the results
rf_clf = RandomForestClassifier(**rf_best_params)
rf_clf.fit(X_train, y_train)

#Print score
print_score(rf_clf, X_train, y_train, X_test, y_test, train=False)

#Plots confusion matrix
y_hat = rf_clf.predict(X_test) > 0.5 
mat = confusion_matrix(y_test, y_hat)
labels = ['No Homicide', 'Homicide']
 
sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=labels, yticklabels=labels)
 
plt.xlabel('Predicted label')
plt.ylabel('True label')
```

In this case, our hyperparameter tuning made our model do a little bit worse than the non-tuned model. Specifically, the results were the except same, except two more individuals were incorrectly predicted to commit a homicide.

There could be a variety of reasons for why this hyperparameter tuning produced suboptimal performance. First, we used a random grid search, which only tests a random subset of parameters. It is possible that the optimal parameters weren't included in this random subset. Replacing our random grid search with a non-random grid search (much like the searches done on the decision tree) would produce better results, albeit it will take longer to run.

Second, we only used a subset of random forest parameters. If we used more, we perhaps could have tested more variables to produce better results.

Third, we specified the search to optimize accuracy. However, a better metric could have been precision or specificity.

Regardless, this hyperparameter example shows that sometimes, hyperparameter tuning is not a cure-all, and it should be done carefully and intentionally.
